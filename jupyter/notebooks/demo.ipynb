# Demo of a Python Jupyter Notebook with Pandas

Import required libraries

```python
from pathlib import Path
import pandas as pd
```

Provide filepath to CSV-file and assert the file exists

```python
filepath: Path = Path(r"../data/titanic.csv")
assert filepath.exists(), f"{filepath.name} not found!"
```

## Show DataFrame

```python
titanic: pd.DataFrame = pd.read_csv(filepath)
titanic
```

## Show Plots of data

Generate a histogram of Fare Price

```python
ax = titanic['Fare'].hist()
title = ax.set_title('Fare Price Distribution')
y_label = ax.set_ylabel('Frequency')
x_label = ax.set_xlabel('Fare Price')
```
